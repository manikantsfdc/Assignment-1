```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
response1 = requests.get('https://www.imdb.com/list/ls056092300/')
```

```python
response1.status_code
#to check the request response
```

```python
html1 = BeautifulSoup(response1.content, 'html.parser')
```

```python
headers = html1.find_all('div', attrs = {'class':'lister-item'} )
```

```python
for header in headers:
    print(header.find('h3').a.text)
    print(header.find('span', attrs = {'class': 'ipl-rating-star__rating'}).text)
    print(header.find_all('span')[1].text)
    print("************************")
```

```python
from openpyxl import Workbook
```

```python
output_sheet_name = 'task3.xlsx'
wb = Workbook()
ws = wb.create_sheet(title='Manikant')
row = ['Movie Name', 'Year', 'Rating']
ws.append(row)

for header in headers:
    a1=header.find('h3').a.text
    a2=header.find('span', attrs = {'class': 'ipl-rating-star__rating'}).text
    a3=header.find_all('span')[1].text
    row = [a1,a3,a2]
    print(a1)
    print(a2)
    print(a3)
    ws.append(row)
    wb.save(output_sheet_name)
```
